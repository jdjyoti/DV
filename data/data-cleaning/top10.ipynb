```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('World_avgTemp.csv')
data['Year'] = data['Year'].astype(str)
data
```

```python
df = data[['Year', 'Month', 'Monthly Anomaly', 'Monthly Unc.', 'Annual Anomaly',
       'Annual Unc.', 'Five-year Anomaly']]
df
```

```python
df = df[df['Year']>= '2000']
df = df[df['Year']< '2022']

df
```

```python
result = df[df.Month == 6].astype(float).nlargest(n=10, columns=['Year','Monthly Anomaly'])
```

```python
result
```

```python
df10 = df[df['Year']> '2011']
df10.rename(columns = {'Monthly Anomaly':'MonthlyAnomaly', 'Annual Anomaly':'AnnualAnomaly', 'Five-year Anomaly': 'FiveYearAnomaly'}, inplace = True)
df10
```

```python
df10.drop(df.columns[[3, 5]], axis=1, inplace=True) 
```

```python
df10
```

```python
compression_opts = dict(method='zip',
                        archive_name='top10.csv')  
df10.to_csv('top10.zip', index=False,
          compression=compression_opts)  
```
